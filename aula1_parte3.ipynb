```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('cursos-prouni.csv',sep=';')
#verificas tamanho do df -> (linhas, coluna)
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
#transformar string em float
df['mensalidade'] = df['mensalidade'].str.replace(',','.')
df['mensalidade'] = df['mensalidade'].astype(float)
df.dtypes
```

```python
df.describe()
```

```python
#descrever dados
df.info()
```

```python
#quantidade de cursos por uf
df['uf_busca'].value_counts()
```

```python
#Qual curso possui maior mensalidade? -> encontrar o maior valor de uma coluna especifica
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
#curso com a menor mensalidade -> encontrar o menor valor de uma coluna especifica
df[df['mensalidade'] == df['mensalidade'].min()]
```

```python
#dataframe[dataframe['coluna'] == dataframe['coluna'].min()][['coluna_x', 'coluna_y']]

#encontrar o menor valor de mais de uma coluna especifica
#-> se for mais de uma coluna criar uma lista composto, se não for, apenas uma lista
df[df['mensalidade'] == df['mensalidade'].min()][['uf_busca', 'cidade_busca']]
```

```python
#mensalidade mais cara por uf
```

```python
#mensalidade mais cara por uf
#dataframe.groupby('campo a ser agrupado')[coluna a ser comparada/consultada].max,min,etc...->  agrupar valores de duas colunas

df.groupby('uf_busca')['mensalidade'].max()
```

```python
#mostrar todos os dados dos cursos mais caros por uf
#dataframe.loc[df.groupby('campo a ser agrupado')[coluna a ser comparada/consultada].idxmax()]
#idxmax -> busca o id da linha e retorna o valor maximo da linha
#loc -> localizar uma linha pelo indice
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()]
```

```python
#mensalidades entre 500 e 1000
df[(df['mensalidade'] > 500) & (df['mensalidade'] < 1000)]
```

```python
# x maiores valores de colunas
#dataframe.nlargest(quantidade de maiores, 'campo'/coluna)
df.nlargest(5, 'mensalidade')
```

```python
# x menores valores de colunas
#dataframe.nsmallest(quantidade de menores, 'campo'/coluna)
df.nsmallest(5, 'mensalidade')
```
